```python
import os
from pprint import pformat
import string

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import flopy

print(f"numpy version: {np.__version__}")
print(f"matplotlib version: {mpl.__version__}")
print(f"flopy version: {flopy.__version__}")
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
#load the model: 
mf = flopy.modflow.Modflow.load(
    "basepfas.nam",
    model_ws=cwd,
    verbose=True,
    exe_name="mfnwt",
    version="mfnwt",
    check=False,
)
```

```python
mt = flopy.mt3d.Mt3dms.load(
    "mtbasepfas.mtnam",
    model_ws=cwd,
    verbose=True,
    version="mt3d-usgs",
    exe_name="mt3dusgs",
    modflowmodel=mf,
)
```

```python
mf.run_model(silent=False, report=True)
```

```python
mt.run_model(silent=False, report=False)
```
