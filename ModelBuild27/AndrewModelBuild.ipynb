# Newton-Raphson Solver: Create a flopy nwt package object

```python
# Newton-Raphson Solver: Create a flopy nwt package object

headtol = 1.0e-4
fluxtol = 5
maxiterout = 5000
thickfact = 1e-06
linmeth = 2
iprnwt = 1
ibotav = 1

nwt = flopy.modflow.ModflowNwt(
    mf,
    headtol=headtol,
    fluxtol=fluxtol,
    maxiterout=maxiterout,
    thickfact=thickfact,
    linmeth=linmeth,
    iprnwt=iprnwt,
    ibotav=ibotav,
    options="SIMPLE",
)
```

```python
# Stress periods [5year, 5year...]; time steps within 1 stress period [1year, 1year, ...]
Steady = np.zeros(40, dtype=bool)
nstp = np.ones(40, dtype = int)*5
tsmult = np.ones(40)*1.0

perlen = np.ones(40)*365*5

nper = len(perlen)
print(nper)

#layercbd: An array of flags indicating whether or not a layer has a Quasi-3D confining bed below it.
laycbd = 0 

# Create the discretization object
# itmuni = 4 (days); lenuni = 2 (meter)
dis = flopy.modflow.ModflowDis(
    mf,
    nlay,
    nrow,
    ncol,
    nper=nper,
    delr=delr,
    delc=delc,
    top=topElev,
    botm=botm,
    laycbd=laycbd,
    itmuni=4,
    lenuni=2,
    steady=Steady,
    nstp=nstp,
    tsmult=tsmult,
    perlen=perlen,
)
```
